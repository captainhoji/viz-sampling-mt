```python
import pandas as pd
import random
from legoBlocks import *
import os
```

```python
def getx():
    n = int(random.random()*40 + 20) #list of length 10 to 100
    x = np.zeros(n)
    for i in range(len(x)-1):
        x[i] = 0.0 + 1/(len(x)-1)*i
    x[-1] = 1.0
    return x

def saveData(x, y, filename):
    d = np.array([x, y]).transpose()
    pd.DataFrame(d).to_csv(f"benchmark/{filename}.csv", header=None, index=None)
```

```python
# Linear
def gen_linear(m, b, noise):
    x = getx()
    y = np.zeros(len(x))
    y = np.array([m*i + b + np.random.normal(0, noise) for i in x])
#     y = (y - y.min()) / (y.max() - y.min())
    return x, y

saveData(*gen_linear(1, 0, .001), 'linear01')
saveData(*gen_linear(-1, 0, .01), 'linear02')
saveData(*gen_linear(1, 0, .1), 'linear03')
```

```python
# Exponential
def gen_exp(a, t, noise):
    x = getx()
    y = np.zeros(len(x))
    y = np.array([a * math.exp(t*i) + np.random.normal(0, noise) for i in x])
#     y = (y - y.min()) / (y.max() - y.min())
    return x, y

saveData(*gen_exp(1, 10, .01), 'exp01')
saveData(*gen_exp(-1, 5, .1), 'exp02')
saveData(*gen_exp(-1, -5, .03), 'exp03')
```

```python
# Spike & Sink
def gen_spike(spike_point, slope, noise, direction):
    x = getx()
    y = np.zeros(len(x)) + np.random.normal(0, noise, len(x))
    x_spike = int(len(x)*spike_point)

    y[x_spike] = direction
    x_slope_start = max(min(int(x_spike - len(x)/slope), x_spike - 1), 0)
    x_slope_end = min(max(int(x_spike + len(x)/slope), x_spike + 1), len(x)-1)

    y[x_slope_start:x_spike] = [direction*i*slope/len(x) + np.random.normal(0, noise) for i in range(x_spike - x_slope_start)]
    y[x_spike+1:x_slope_end] = [direction*(x_slope_end - x_spike - i)*slope/len(x) + np.random.normal(0, noise) for i in range(x_slope_end - x_spike - 1)]

    return x, y

saveData(*gen_spike(0.2, 10, 0.1, 1), 'spike01')
saveData(*gen_spike(0.7, 5, 0.1, -1), 'spike02')
saveData(*gen_spike(0.5, 2, 0.1, 1), 'spike03')
```

```python
# Rise & Drop
def gen_rise(slope, direction, noise):
    x = getx()
    y = np.zeros(len(x))
    
    x_slope_start = int(len(x)/2 - len(x)/slope)
    x_slope_end = int(len(x)/2 + len(x)/slope)
    
    y[x_slope_start:x_slope_end] = [min(max(direction * i * slope/(2*len(x)), -1), 1) for i in range(x_slope_end - x_slope_start)]
    y[x_slope_end:] = [direction for i in range(len(y[x_slope_end:]))]
    y += np.random.normal(0, noise, len(x))
    return x, y

saveData(*gen_rise(100, 1, 0.1), 'rise01')
saveData(*gen_rise(20, -1, 0.1), 'rise02')
saveData(*gen_rise(5, 1, 0.01), 'rise03')
```

```python
# Plateau
def gen_plateau(rise_center, slope, duration, direction, noise):
    x = getx()
    y = np.zeros(len(x))
    x_flat_start = int(len(x)*(rise_center-duration/2))
    x_flat_end = int(len(x)*(rise_center + duration/2))

    y[x_flat_start:x_flat_end] = direction
    x_slope_start = max(min(int(x_flat_start - len(x)/slope), x_flat_start - 1), 0)
    x_slope_end = min(max(int(x_flat_end + len(x)/slope), x_flat_end + 1), len(x)-1)

    y[x_slope_start:x_flat_start] = [direction*i*slope/len(x) for i in range(x_flat_start - x_slope_start)]
    y[x_flat_end:x_slope_end] = [direction*(x_slope_end - x_flat_end - i-1)*slope/len(x) for i in range(x_slope_end - x_flat_end)]
    y += np.random.normal(0, noise, len(x))
    return x, y

saveData(*gen_plateau(0.5, 100, 0.6, 1, 0.1), 'plateau01')
saveData(*gen_plateau(0.3, 20, 0.2, -1, 0.1), 'plateau02')
saveData(*gen_plateau(0.5, 5, 0.35, 1, 0.05), 'plateau03')
```

```python
# Periodic
def gen_periodic_sin(period, noise):
    x = getx()
    y = [math.sin(i/period) for i in x]
    y += np.random.normal(0, noise, len(x))
    return x, y

def gen_periodic_box(period, noise):
    x = getx()
    y = np.zeros(len(x))
    period_halflen = int(len(x)*period/2)
    switch = True
    y[0], y[1] = 1, 1
    j = 0
    for i in range(2, len(x)):
        if j == period_halflen:
            j=0
            switch = not switch
        if switch:
            y[i] = 0
        else:
            y[i] = 1
        j+=1
    y += np.random.normal(0, noise, len(x))
    return x, y

saveData(*gen_periodic_sin(0.06, 0.1), 'periodic01')
saveData(*gen_periodic_box(0.25, 0.01), 'periodic02')
# saveData(*gen_periodic_sin(0.02, 0.01), 'periodic03')
```
